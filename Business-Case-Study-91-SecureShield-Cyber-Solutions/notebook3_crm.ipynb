#  SecureShield: CRM & Business Insights

Analysis of client satisfaction (CSAT) and renewal probabilities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
df = pd.read_csv("secureshield_crm.csv")
```

## 1. CSAT scores by Industry

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="industry", y="satisfaction_score", hue="industry", palette="mako", errorbar=None, legend=False)
plt.title("Mean Client Satisfaction Score (CSAT) by Sector")
plt.ylabel("Average CSAT Score")
plt.show()
```

**Explanation:** This bar graph compares average customer satisfaction scores, showing high service quality across all 8 industry sectors.

## 2. Vulnerability Scan Score Distribution

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="industry", y="vulnerability_scan_score", hue="industry", palette="magma", errorbar=None, legend=False)
plt.title("Mean Vulnerability Scan Results across Industries")
plt.ylabel("Average Vulnerability Score")
plt.show()
```

**Explanation:** This chart shows the average security health of our clients, confirming that all industries maintain safe vulnerability scores.

## 3. Financial Health: Renewal Probability Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["renewal_probability"], bins=20, kde=True, color="#00cc96")
plt.title("Distribution of Client Renewal Probability")
plt.xlabel("Probability of Renewal (0.0 - 1.0)")
plt.ylabel("Number of Clients")
plt.show()
```

**Explanation:** The histogram confirms a healthy recurring revenue model, with most clients showing a 90% or higher probability of renewal.
